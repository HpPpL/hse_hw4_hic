## Выбираем участок и начинаем обработку

```python
chromosome = "2L"
start = 5000000  # Начальная позиция участка
end = 6500000    # Конечная позиция участка
region = f"{chromosome}:{start}-{end}"
```

```python
import cooler
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import cooltools
from cooltools import insulation


# Шаг 1-2: Загрузка файлов HiC с указанными разрешениями
file_paths = [
    "/content/HiC1.dm3.mapq_30.1000.mcool",
    "/content/HiC2.dm3.mapq_30.1000.mcool",
    "/content/HiC3.dm3.mapq_30.1000.mcool",
    "/content/HiC4.dm3.mapq_30.1000.mcool"
]

# Шаг 4: Информация о Hi-C матрице для каждого файла
for i, file_path in enumerate(file_paths):
    clr = cooler.Cooler(file_path + '::/resolutions/10000')  # Указание разрешения 10000

    print(f"\nInformation about HiC{1+i} Cooler file:")
    for key, value in clr.info.items():  # Печать информации о файле
        print(f"{key}: {value}")

    # Шаг 5 и 6: Работа с балансированной матрицей для выбранного участка хромосомы
    chromosome = 'chr2L'
    start, end = 5000000, 6500000
    if chromosome in clr.chromnames:
        matrix = clr.matrix(balance=True).fetch(f'{chromosome}:{start}-{end}')
        plt.figure(figsize=(10, 8))
        plt.title(f'Balanced Contact Matrix for HiC{1+i} on {chromosome}:{start}-{end}')
        plt.imshow(np.log1p(matrix), cmap='YlOrRd', interpolation='none')
        plt.colorbar(label='Log1p(contact count)')
        plt.show()

        # Шаг 9-11: Расчет insulation score и анализ TAD границ
        windows = [500000, 1000000]  # Размеры окон в базах для расчета
        insulation_score = insulation(clr, windows, verbose=True)
        print(f"Insulation Score for HiC{1+i}:\n", insulation_score.head())

        # Визуализация insulation score
        fig, ax = plt.subplots(figsize=(10, 4))
        ax.plot(insulation_score['start'], insulation_score['log2_insulation_score_500000'], label='500 kb window')
        ax.plot(insulation_score['start'], insulation_score['log2_insulation_score_1000000'], label='1 Mb window')
        ax.set_title('Insulation Score over genomic bins')
        ax.set_xlabel('Genomic position (bp)')
        ax.set_ylabel('Log2 Insulation Score')
        ax.legend()
        plt.show()

        # Шаг 12: Создание BED файлов для TAD границ
        boundaries = insulation_score.loc[insulation_score['is_boundary_1000000']].copy()
        boundaries['score'] = boundaries['boundary_strength_1000000']
        boundaries.to_csv(f'/content/TAD_boundaries_HiC{1+i}.bed', sep='\t', index=False, header=False, columns=['chrom', 'start', 'end', 'score'])
```
